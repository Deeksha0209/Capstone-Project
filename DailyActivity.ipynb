#Description of Dataset -

This dataset encompasses Fitbit users' fitness tracking data spread across multiple datasets and columns. The "DailyActivity" dataframe includes data on total steps, tracker distance, types of distances, levels of activity minutes, and calories for various users across different days and time intervals. # **#Let's Begin the Data Cleaning Process!**

```python
#Let's check and display duplicated values, if any.
df.duplicated().value_counts()
#Great, there is no duplicacy in the df.
```

```python
#Let's take a look at the null values, if any!
df.isna().sum()
#Awesome! No null values. This makes our data cleaning easier.
```

```python
#ActivityDate column has object data type. I shall set the correct data type for this colm.
#dt.day_name() function will display the weekdays associated with the respective date in a separate colm.
df['ActivityDate'] = pd.to_datetime(df['ActivityDate'])
df['Weekday'] = df['ActivityDate'].dt.day_name()
```

# **#Lets begin EDA process!**

```python
#Lets check the no of unique ids!
df['Id'].nunique()
```

```python
#Checking the minimum and maximum values for calories colm
minimum_cal = df['Calories'].min()
print("Minimum Calories:", minimum_cal)
maximum_cal = df['Calories'].max()
print("Maximum Calories:", maximum_cal)
```

```python
#Checking the minimum & maximum values for Total Steps!
minimum_steps = df['TotalSteps'].min()
print("Minimum Total Steps:", minimum_steps)
maximum_steps = df['TotalSteps'].max()
print("Maximum Total Steps:", maximum_steps)
```

```python
#Checking the median & max stats for Tracker Distance colm
median_trackdistance = df['TrackerDistance'].median()
print("Median Tracker Distance:", median_trackdistance)
maximum_distance = df['TotalDistance'].max()
print("Maximum Total Distance:", maximum_distance)
```

### **#Bivariate & Multivariate Analysis**

```python
#Plotting scatter plot for Distances!
from matplotlib import pyplot as plt
id_df.plot(kind='scatter', x='VeryActiveDistance', y='ModeratelyActiveDistance', s=25, alpha=1, color='green')
plt.title('Scatter Plot of Distances', fontsize=14)
plt.show()
```

```python
#Let's create a scatter plot for the Total Steps and the the Total Distance covered by the user
from matplotlib import pyplot as plt
df.plot(kind='scatter', x='TotalSteps', y='TotalDistance', s=25, alpha=1, color="magenta")
plt.title('Total Steps v/s Total Distance', fontsize=14)
plt.show()
#Wow, the scatter points are increasing linearly majorly lying between 0-15000 total steps.
```

```python
#Creating a line chart for Total Distance v/s Tracker Distance covered by users!
plt.figure(figsize=(8, 6))
sns.set_palette("deep")
sns.lineplot(x='TotalDistance', y='TrackerDistance', data=df)
plt.xlabel('TotalDistance', fontsize=12)
plt.ylabel('TrackerDistance', fontsize=12)
plt.title('Line Plot of TotalDistance v/s TrackerDistance', fontsize=14)
plt.show()
```

```python
#Let's plot another line chart depicting The tracker Distance entries v/s Acitive Minutes.
plt.figure(figsize=(8,6))
sns.set_palette("cubehelix")
sns.lineplot(x='TrackerDistance', y='VeryActiveMinutes', data=df)
plt.xlabel('TrackerDistance', fontsize=12)
plt.ylabel('VeryActiveMinutes', fontsize=12)
plt.title('Line Plot of TrackerDistance v/s VeryActiveMinutes', fontsize=14)
plt.show()
```

```python
#We also have a weekdays columns. Lets plot a bar chart using that wrt the Total distance covered by the users!
plt.figure(figsize=(10, 6))
sns.barplot(x='Weekday', y='TotalDistance', data=df, palette="dark")
plt.xlabel('Weekdays')
plt.ylabel('Total Distance')
plt.title('Total Distance covered on Weekdays')
plt.show()
#Wow, looks like Saturday was the day when max total distance was covered, closely followed by Tuesday & Monday.
```

```python
#Lets depict a pair plot relationship among Calories, Total steps & Total Distance.
sns.pairplot(df[['TotalSteps', 'TotalDistance', 'Calories']])
plt.show()
```

```python
#Heatmaps are best way to understand the correlation among numerical columns! Heatmap values lie between -1 to +1.
numerical_columns = ['VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryActiveDistance']
correlation_matrix = df[numerical_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="viridis")
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

```python
#Heatmaps are best way to understand the correlation among numerical columns! Heatmap values lie between -1 to +1.
numerical_columns = ['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes']
correlation_matrix = df[numerical_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="rocket")
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

```python
#Let's take one final plot and complete our EDA - A cluster map using seaborn. It is basically a clustered heatmap for numerical columns. A heatmap, as we know, represents the correlation matrix b/w columns.
#Cluster map refers to hierarchal clustering applied to rows and columns of a hatmap based on their similarity in terms of correlation. the colors here represent the strength of correlation.
#A darker color indicates stronger correlation( can be either +ve or -ve )!
numerical_columns = ['TrackerDistance', 'TotalSteps', 'Calories']
plt.figure(figsize=(10,6))
sns.clustermap(df[numerical_columns].corr(), annot=True, cmap='flare')
plt.show()
```
