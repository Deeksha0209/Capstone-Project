# **#Merging of Datasets**

```python
#Now that the data cleaning for each df is complete, we shall go ahead and merge the Dfs.
merged_df = pd.merge(df1, df2, on='ActivityHour', how='inner')
merged_df.drop(['Id_y'], axis=1, inplace=True)  # Drop 'Id_df2' column
merged_df.rename(columns={'Id_x': 'Id'}, inplace=True)
merged_df
```

```python
#We now have our final df - hourlyActivity_df !
hourlyActivity_df = pd.merge(merged_df, df3, on='ActivityHour', how='inner')
hourlyActivity_df.drop(['Id_y'], axis=1, inplace=True)  # Drop 'Id_df2' column
hourlyActivity_df.rename(columns={'Id_x': 'Id'}, inplace=True)
hourlyActivity_df
```

```python
#Lets take a look at the shape of our final df
hourlyActivity_df.shape
```

```python
#Lets do one final chk for duplicate values in our merged df!
hourlyActivity_df.duplicated().value_counts()
```

```python
#Dropping the duplicate rows to clean our data!
hourlyActivity_df=hourlyActivity_df.drop_duplicates()
```

```python
hourlyActivity_df['ActivityHour'] = pd.to_datetime(hourlyActivity_df['ActivityHour'])
hourlyActivity_df['Weekday'] = hourlyActivity_df['ActivityHour'].dt.day_name()
```

```python
#Shall we take one last look at the null values, if any!
hourlyActivity_df.isna().sum()
#Great, now our Data cleaning is complete.
```

```python
#Lets take one last look at the updated shape of our final df -
hourlyActivity_df.shape
```

# **#Let's Begin EDA Process on Hourly Data!!**

**#Univariate Analysis**

```python
#Lets display values for a particular user
id_df=hourlyActivity_df.loc[hourlyActivity_df['Id'] == 4558609924]
id_df
```

```python
#Plotting the Distribution plot for Step Total values
plt.figure(1)
sns.distplot(hourlyActivity_df['StepTotal'])
plt.xlabel('StepTotal', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.title('Distribution of Step Total Using distplot', fontsize=14)
plt.show()
```

```python
#Creating a 1D scatter plot for Step Total
plt.plot(id_df['StepTotal'],np.zeros_like(id_df['StepTotal']),'p')
plt.show()
```

```python
#Creating a 1D scatter plot for Calories
plt.plot(id_df['Calories'],np.zeros_like(id_df['Calories']),'p')
plt.show()
```

```python
#Lets take a ook at the kernal density plot for Average Intensity
sns.kdeplot(data=hourlyActivity_df['AverageIntensity'],label='AverageIntensity', shade=True)
```

# **#BiVariate & MultiVariate Analysis!**

```python
#Creating a scatter plot for the Total Intensity & Calories for a particular user
id_df.plot(kind='scatter', x='TotalIntensity', y='Calories', s=25, alpha=0.5, color="orange")
plt.title('Scatter Plot of Total Intensity v/s Calories', fontsize=14)
plt.show()
```

```python
#Displaying a line plot for Total steps v/s Avg Intensity of our entire df!
plt.figure(figsize=(8,6))
sns.set_palette("cubehelix")
sns.lineplot(x='StepTotal', y='AverageIntensity', data=hourlyActivity_df)
plt.xlabel('Step Total', fontsize=12)
plt.ylabel('Average Intensity', fontsize=12)
plt.title('Line Plot of Step Total v/s Average Intensity', fontsize=14)
plt.show()
```

```python
#Let's recall the entries for our user ( which we have displayed above )
id_df.head(3)
```

```python
#Displaying a bubble chart for Total steps & Total intensity for a particular user id!
x = id_df['StepTotal']  # X-axis data
y = id_df['TotalIntensity']  # Y-axis data
bubble_size = id_df['AverageIntensity']  # Bubble size data

plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=bubble_size, alpha=0.5)  # 's' parameter controls the size of bubbles
plt.xlabel('Step Total')
plt.ylabel('Total Intensity')
plt.title('Bubble Chart')
plt.colorbar(label='Bubble Size')  # Adding color bar for bubble sizes
plt.grid(True)
plt.show()
```

```python
#Creating a bar chart for our user 4558609924 to display the weekday wherein the user burnt max & min calories!
plt.figure(figsize=(10, 6))
sns.barplot(x='Weekday', y='Calories', data=id_df, palette="viridis")
plt.xlabel('Weekdays')
plt.ylabel('Total Calories')
plt.title('Total Calories on Weekdays')
plt.show()

#Wow, user 4558609924 burnt max calories on Sunday closely followed by Thursday & the least calories on Wednesday.
```

```python
#Now let's gather some insights for another user!
id_df=hourlyActivity_df.loc[hourlyActivity_df['Id'] == 1503960366]
id_df
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Weekday', y='StepTotal', data=id_df, palette="magma")
plt.xlabel('Weekdays')
plt.ylabel('Total Step')
plt.title('Total Steps Walked on Weekdays by User')
plt.show()
#Wow, looks like Saturdays are the best days for max steps for this user!
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='AverageIntensity', y='Weekday', data=id_df, palette="icefire", orient='h')
plt.xlabel('Average Intensity')
plt.ylabel('Weekdays')
plt.title('Average Intensity of user on Weekdays')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='TotalIntensity', y='Weekday', data=id_df, palette="dark", orient='h')
plt.xlabel('Total Intensity')
plt.ylabel('Weekdays')
plt.title('Total Intensity of user on Weekdays')
plt.show()
```

```python
#Creating a correlation heatmap for total steps, intensity & calories. Correlation matrix ranges from -1 to +1 .
numerical_columns = ['StepTotal', 'AverageIntensity', 'TotalIntensity', 'Calories']
correlation_matrix = hourlyActivity_df[numerical_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="viridis")
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```
